```python
import os
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET
finallist=[]
grandfinale=[]
path = "cctv"
for filename in os.listdir(path):
    pa="cctv\\" + filename
    for file in os.listdir(pa):
        f="cctv\\" + filename + "\\" + file
        tree=ET.parse(f)
        root= tree.getroot()
        part000list=[]
        for child in root:
            part000list.append(child.text)
        finallist=finallist+part000list
    grandfinale=grandfinale+finallist
len(grandfinale)
```

```python
import os

path = "cctvraw"
with open('cctv.txt', 'w') as outfile:
    for fname in os.listdir(path):
        fd="cctvraw\\" + fname
        with open(fd) as infile:
            for line in infile:
                outfile.write(line)
```

```python
main_list=[]
with open('cctv.txt') as infile:
    for line in infile:
        if line!='\n' and line[0]!='#':
            split_line=line.split()
            main_list.append(split_line)
            
    df=pd.DataFrame(main_list)
df.to_csv('cctv.csv',index=False)
```
